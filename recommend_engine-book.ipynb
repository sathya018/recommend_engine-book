```python
book_df = book.rename({'Unnamed: 0':'indx','User.ID':'userid','Book.Title':'book_title','Book.Rating':'book_rating'},axis =1)
```

```python
user_book_df =book_df.pivot_table(index='userid',
                                 columns='book_title',
                                 values='book_rating').reset_index(drop=True)
```

```python
user_book_df
```

```python
user_book_df.index = book_df.userid.unique()
```

```python
user_book_df
```

```python
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book_df.userid.unique()
user_sim_df.columns = book_df.userid.unique()
```

```python
user_sim_df.iloc[0:20, 0:20]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:125]
```
